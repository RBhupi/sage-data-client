# Basic Plotting Example

```python
import sage_data_client
```

First, we'll query the last 7 days of temperature data from W022's BME680 sensor.

```python
df = sage_data_client.query(
    start="-7d",
    filter={
        "name": "env.temperature",
        "vsn": "W022",
        "sensor": "bme680",
    }
)
```

Next, we'll plot a simple line chart.

```python
df.set_index("timestamp").value.plot()
```

Finally, we'll plot the temperature distribution.

```python
df.value.hist(bins=100)
```
